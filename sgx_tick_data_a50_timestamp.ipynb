```python
order_book_bid = []
order_book_ask = []
bid = []
ask = []
timestamp_ = []
for i in range(0,len(data.timestamp.unique()),1):#,len(data.timestamp.unique()),1):
    x = data.timestamp.unique()[i]
    timestamp_.append(x)
    y = data[data['timestamp'].str.contains(x)]
    bid.append(y[(y.bid_or_ask == 1)][["price","quantity_difference"]]) # bid
    ask.append(y[(y.bid_or_ask == 2)][["price","quantity_difference"]]) # ask
    if i == 0:
        a = bid[i].sort(['price'],ascending = [True])
        b = ask[i].sort(['price'],ascending = [True])
    elif i > 0:
        a = pd.concat([order_book_bid[i-1],bid[i]]).groupby(by=['price'],as_index = False).sum().sort(['price'],ascending = [False])
        b = pd.concat([order_book_ask[i-1],ask[i]]).groupby(by=['price'],as_index = False).sum().sort(['price'],ascending = [True])
    order_book_bid.append(a[a.quantity_difference != 0])
    order_book_ask.append(b[b.quantity_difference != 0])
    
```

```python
import time 

start = time.time()

order_book_bid = []
order_book_ask = []
bid = []
ask = []
order_number_bid = []
order_number_ask = []
timestamp_ = []
for i in range(0,len(data.timestamp.unique()),1):
    x = data.timestamp.unique()[i]
    timestamp_.append(x)
    y = data[data['timestamp'].str.contains(x)]
    bid.append(y[(y.bid_or_ask == 1)][["price","quantity_difference"]]) # bid
    ask.append(y[(y.bid_or_ask == 2)][["price","quantity_difference"]]) # ask
    order_number_bid.append(y[(y.bid_or_ask == 1)][["order_number"]])
    order_number_ask.append(y[(y.bid_or_ask == 2)][["order_number"]])
    if i == 0:
        a = bid[i].sort(['price'],ascending = [True])
        b = ask[i].sort(['price'],ascending = [True])
    elif i > 0:
        a = pd.concat([order_book_bid[i-1],bid[i]]).groupby(by=['price'],as_index = False).sum().sort(['price'],ascending = [False])
        b = pd.concat([order_book_ask[i-1],ask[i]]).groupby(by=['price'],as_index = False).sum().sort(['price'],ascending = [True])
    order_book_bid.append(a[a.quantity_difference != 0])
    order_book_ask.append(b[b.quantity_difference != 0])
end = time.time()  
print "Total time = %f"%(end - start)
```
